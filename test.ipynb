```python
from rdkit import Chem
from rdkit.Chem import Draw

from ReactiveSiteFeatures import Local_FromSmiles, Global_FromSmiles, Local_FromInchi, Global_FromInchi
```

```python
smi = 'NC1CPNC(N)=C1C1=CN=CC=C1'
inchi = Chem.MolToInchi(Chem.MolFromSmiles(smi))
```

```python
def mol_with_atom_index(mol):
    for atom in mol.GetAtoms():
        atom.SetAtomMapNum(atom.GetIdx())
    return mol

# draw mol based on original smiles string
mol = mol_with_atom_index(Chem.MolFromSmiles(smi))
idx = 6
Chem.Draw.MolToImage(mol, highlightAtoms=[idx])
```

```python
# draw mol based on inchi, note that indices differ
mol = mol_with_atom_index(Chem.MolFromInchi(inchi))
idx2 = 10
Chem.Draw.MolToImage(mol, highlightAtoms=[idx2])
```

## Local features based on SMILES, radius 2

```python
results = Local_FromSmiles(smiles=smi, index=idx, radius=2)
print(results)
```

## Local features based on INCHI, radius 2

```python
results = Local_FromInchi(inchi=inchi, index=idx2, radius=2)
print(results)
```

## Local features based on SMILES, radius 6

```python
results = Local_FromSmiles(smiles=smi, index=idx, radius=6)
print(results)
```

## Local features based on INCHI, radius 6

```python
results = Local_FromInchi(inchi=inchi, index=idx2, radius=6)
print(results)
```

## Global features based on SMILES

```python
results = Global_FromSmiles(smiles=smi)
print(results)
```

## Global features based on INCHI

```python
results = Global_FromInchi(inchi=inchi)
print(results)
```
